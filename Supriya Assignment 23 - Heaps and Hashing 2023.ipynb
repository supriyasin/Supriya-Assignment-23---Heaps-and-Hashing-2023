```python
#<aside>
ðŸ’¡ Question-1:

Given preorder of a binary tree, calculate itsÂ **[depth(or height)](https://www.geeksforgeeks.org/write-a-c-
program-to-find-the-maximum-depth-or-height-of-a-tree/)**Â [starting from depth 0]. The preorder is given as
a string with two possible characters.

1. â€˜lâ€™ denotes the leaf
2. â€˜nâ€™ denotes internal node

The given tree can be seen as a full binary tree where every node has 0 or two children. The two children of a
node can â€˜nâ€™ or â€˜lâ€™ or mix of both.

**Examples :**

Input  : nlnll
Output : 2
Explanation :

!https://media.geeksforgeeks.org/wp-content/uploads/btree1.png

Input  : nlnnlll
Output : 3

</aside>


"""To calculate the depth (or height) of a binary tree from its given preorder representation, you can use a 
   recursive approach. The key is to keep track of the number of 'n' characters encountered before finding the 
   corresponding 'l' character, as this indicates the depth of the tree at that point."""

# Let's write a Python function to calculate the depth:

def find_depth(preorder):
    def helper(index):
        if index >= len(preorder):
            return 0, index

        if preorder[index] == 'l':
            return 0, index

        left_depth, next_index = helper(index + 1)
        right_depth, final_index = helper(next_index + 1)

        current_depth = 1 + max(left_depth, right_depth)
        return current_depth, final_index

    depth, _ = helper(0)
    return depth

# Test cases
print(find_depth("nlnll"))    # Output: 2
print(find_depth("nlnnlll"))  # Output: 3

"""Explanation of the function:

   1. We define a helper function helper(index) that takes an index and returns the depth of the tree from
      that index onwards and the next index to be processed.
      
   2. The function first checks if the index is out of bounds. If so, it returns a depth of 0 and the same index.
   
   3. If the character at the current index is 'l', it means we have reached a leaf node, and the depth is 0. 
      It returns a depth of 0 and the same index to indicate that no more nodes are present at this level.
      
   4. If the character at the current index is 'n', it means we have encountered an internal node. We recursively
      call the helper function to find the depth of the left subtree and the next index to be processed.
      
   5. Then, we recursively call the helper function again to find the depth of the right subtree and the final
      index after processing both left and right subtrees.
      
   6. Finally, we calculate the current depth as 1 plus the maximum of the depths of the left and right subtrees 
      and return it along with the final index.
      
   The function find_depth is called with the given preorders, and it returns the expected depth for each case."""
```

```python
#<aside>
ðŸ’¡ Question-2:

Given aÂ Binary tree, the task is to print theÂ **left view**Â of the Binary Tree. The left view of a Binary 
Tree is a set of leftmost nodes for every level.

**Examples:**

***Input:***

            4

          / Â  \

        5 Â  Â  2

             / Â  \

            3 Â  Â  1

           / Â \

          6 Â  Â 7

***Output:**Â 4 5 3 6*

**Explanation:**

!https://media.geeksforgeeks.org/wp-content/cdn-uploads/left-view.png

***Input:***

                    1

                  / Â  \

                2 Â  Â  Â  3

                 \

                   4

                     \

                        5

                           \

                             6

**Output:**Â 1 2 4 5 6

</aside>

"""To print the left view of a binary tree, you can perform a level order traversal and print the first node 
   encountered at each level."""

#Let's implement this in Python:


class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def left_view(root):
    if not root:
        return

    queue = [(root, 1)]
    current_level = 0

    while queue:
        node, level = queue.pop(0)

        if level > current_level:
            print(node.val, end=" ")
            current_level = level

        if node.left:
            queue.append((node.left, level + 1))

        if node.right:
            queue.append((node.right, level + 1))


# Test cases
# Tree 1
root1 = TreeNode(4)
root1.left = TreeNode(5)
root1.right = TreeNode(2)
root1.right.left = TreeNode(3)
root1.right.right = TreeNode(1)
root1.right.left.left = TreeNode(6)
root1.right.left.right = TreeNode(7)
print("Left view of Tree 1:", end=" ")
left_view(root1)  # Output: 4 5 3 6

# Tree 2
root2 = TreeNode(1)
root2.left = TreeNode(2)
root2.right = TreeNode(3)
root2.left.right = TreeNode(4)
root2.left.right.right = TreeNode(5)
root2.left.right.right.right = TreeNode(6)
print("\nLeft view of Tree 2:", end=" ")
left_view(root2)  # Output: 1 2 4 5 6

"""Explanation of the code:

   1. We define a TreeNode class to represent the nodes of the binary tree.
   
   2. The left_view function takes the root node of the binary tree as input and prints the left view of the 
      tree using a level order traversal approach.
      
   3. We use a queue to perform the level order traversal. Each element in the queue is a tuple containing the 
      node and its level in the tree.
      
   4. We initialize the current_level variable to 0 to keep track of the current level being processed. 
      We also start the traversal from the root node with level 1.
      
   5. During the traversal, when we encounter a node at a level greater than the current_level, we print its
      value (as it's the leftmost node at that level) and update the current_level.
      
   6. We continue traversing the tree by adding the left and right children of each node to the queue along
      with their respective levels.
      
   7. The traversal continues until the queue is empty, and we have processed all nodes in the tree."""
```

```python
#<aside>
ðŸ’¡ Question-3:

Given a Binary Tree, print the Right view of it.

TheÂ right view of a Binary TreeÂ is a set of nodes visible when the tree is visited from the Right side.

**Examples:**

**Input:**

         1

      / Â  Â  \

   2 Â  Â   Â  Â 3

/ Â  \ Â  Â  Â  / Â \

4 Â  Â  5 Â  6 Â  Â 7

             \

               8

**Output**: 

Right view of the tree is 1 3 7 8

**Input:**

         1

       /

    8

  /

7

**Output**: 

Right view of the tree is 1 8 7

</aside>

"""To print the right view of a binary tree, you can perform a level order traversal and print the last node
   encountered at each level. This will give you the nodes visible from the right side."""

#Let's implement this in Python:

class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def right_view(root):
    if not root:
        return

    queue = [(root, 1)]
    current_level = 0
    rightmost_node = None

    while queue:
        node, level = queue.pop(0)

        if level > current_level:
            if rightmost_node:
                print(rightmost_node.val, end=" ")
            current_level = level
            rightmost_node = None

        rightmost_node = node

        if node.left:
            queue.append((node.left, level + 1))

        if node.right:
            queue.append((node.right, level + 1))

    if rightmost_node:
        print(rightmost_node.val, end=" ")


# Test cases
# Tree 1
root1 = TreeNode(1)
root1.left = TreeNode(2)
root1.right = TreeNode(3)
root1.left.left = TreeNode(4)
root1.left.right = TreeNode(5)
root1.right.left = TreeNode(6)
root1.right.right = TreeNode(7)
root1.right.right.right = TreeNode(8)
print("Right view of Tree 1:", end=" ")
right_view(root1)  # Output: 1 3 7 8

# Tree 2
root2 = TreeNode(1)
root2.left = TreeNode(8)
root2.left.left = TreeNode(7)
print("\nRight view of Tree 2:", end=" ")
right_view(root2)  # Output: 1 8 7

"""Explanation of the code:

   1. We define a TreeNode class to represent the nodes of the binary tree.
   
   2. The right_view function takes the root node of the binary tree as input and prints the right view of the 
      tree using a level order traversal approach.
      
   3. We use a queue to perform the level order traversal. Each element in the queue is a tuple containing the
      node and its level in the tree.
      
   4. We initialize the current_level variable to 0 to keep track of the current level being processed.
      We also start the traversal from the root node with level 1.
      
   5. During the traversal, when we encounter a node at a level greater than the current_level, we print the 
      value of the rightmost_node (as it's the rightmost node at the previous level) and update the current_level.
      
   6. We continue traversing the tree by adding the left and right children of each node to the queue along with
      their respective levels.
      
   7. The traversal continues until the queue is empty, and we have processed all nodes in the tree. 
      The last node encountered at each level is considered part of the right view.
      
   8. After the traversal, if there is any rightmost_node left, we print its value as it will be the rightmost
      node at the deepest level."""
```

```python
#<aside>
ðŸ’¡ Question-4:

Given a Binary Tree, The task is to print theÂ **bottom view**Â from left to right. A nodeÂ **x**Â is there in output
if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is
equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1.

**Examples:**

**Input:**

             20

           / Â  Â  \

        8 Â  Â  Â  Â  22

    / Â  Â  Â \ Â  Â  Â  Â  \

5 Â  Â  Â  Â  3 Â  Â  Â  Â 25

        / Â  Â \

   10 Â  Â  Â  14

**Output:**Â 5, 10, 3, 14, 25.

**Input:**

             20

           / Â  Â  \

        8 Â  Â  Â  Â  22

    / Â  Â  Â \ Â  Â  Â / Â  \

 5 Â  Â  Â  Â  3 Â   4 Â  Â  25

         / Â  Â \

     10 Â  Â  Â  14

**Output:**

5 10 4 14 25.

**Explanation:**

If there are multiple bottom-most nodes for a horizontal distance from the root, then print the later one in the 
level traversal.

**3 and 4** are both the bottom-most nodes at a horizontal distance of 0, we need to print 4.

</aside>

"""To print the bottom view of a binary tree, you can perform a level order traversal while keeping track of the
   horizontal distance of each node from the root. For each horizontal distance, you update the value of the node 
   encountered at that distance. This way, the last node encountered at each horizontal distance will be the 
   bottommost node for that distance."""

#Let's implement this in Python:

from collections import deque

class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def bottom_view(root):
    if not root:
        return

    queue = deque([(root, 0)])
    distance_map = {}

    while queue:
        node, hd = queue.popleft()

        distance_map[hd] = node.val

        if node.left:
            queue.append((node.left, hd - 1))

        if node.right:
            queue.append((node.right, hd + 1))

    # Print the nodes from the distance_map in ascending order of horizontal distance
    for hd in sorted(distance_map.keys()):
        print(distance_map[hd], end=" ")


# Test cases
# Tree 1
root1 = TreeNode(20)
root1.left = TreeNode(8)
root1.right = TreeNode(22)
root1.left.left = TreeNode(5)
root1.left.right = TreeNode(3)
root1.right.right = TreeNode(25)
root1.left.right.left = TreeNode(10)
root1.left.right.right = TreeNode(14)
print("Bottom view of Tree 1:")
bottom_view(root1)  # Output: 5 10 3 14 25

# Tree 2
root2 = TreeNode(20)
root2.left = TreeNode(8)
root2.right = TreeNode(22)
root2.left.left = TreeNode(5)
root2.left.right = TreeNode(3)
root2.right.left = TreeNode(4)
root2.right.right = TreeNode(25)
root2.left.right.left = TreeNode(10)
root2.left.right.right = TreeNode(14)
print("\nBottom view of Tree 2:")
bottom_view(root2)  # Output: 5 10 4 14 25

"""Explanation of the code:

   1. We define a TreeNode class to represent the nodes of the binary tree.
   
   2. The bottom_view function takes the root node of the binary tree as input and prints the bottom view of 
      the tree using a level order traversal approach.
      
   4. We use a deque to perform the level order traversal. Each element in the deque is a tuple containing the 
      node and its horizontal distance (hd) from the root.
      
   5. We initialize a distance_map dictionary to store the value of the last node encountered at each horizontal distance.
   
   6. During the traversal, we update the value of the distance_map with the value of the node encountered at the current 
      horizontal distance. Since we are using a deque for the traversal, the last node encountered at each horizontal 
      distance will overwrite the previous ones, effectively giving us the bottommost node for that distance.
      
   7. We continue traversing the tree by adding the left and right children of each node to the deque along with their
      respective horizontal distances.
      
   8. After the traversal is complete, we print the values of the nodes from the distance_map in ascending order of 
      horizontal distance, which gives us the bottom view of the tree."""
```
